```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

np.random.seed(42)
random.seed(42)

start_id = 21
num_rows = 600

departments = ["IT", "Sales", "Finance", "HR", "Operations"]

roles = {
    "IT": ["Software Engineer", "Data Analyst", "Data Engineer", "Business Analyst", "Software Developer"],
    "Sales": ["Sales Executive", "Sales Manager", "Sales Associate"],
    "Finance": ["Accountant", "Financial Analyst", "Auditor", "Treasury Analyst"],
    "HR": ["HR Manager", "HR Executive", "Recruiter", "HR Coordinator"],
    "Operations": ["Team Lead", "Supervisor", "Logistics Manager", "Operations Analyst"]
}

education_map = {
    "IT": ["B.Tech", "B.Sc", "M.Tech"],
    "Sales": ["BBA", "MBA"],
    "Finance": ["B.Com", "MBA"],
    "HR": ["BA", "MBA"],
    "Operations": ["BBA", "MBA"]
}

locations = ["Bangalore", "Mumbai", "Delhi", "Hyderabad", "Pune"]

data = []

for i in range(num_rows):
    emp_id = start_id + i
    gender = random.choice(["Male", "Female"])
    name = f"Employee_{emp_id}"   # simple generated name
    department = random.choice(departments)
    role = random.choice(roles[department])
    salary = random.randint(40000, 100000)
    
    hire_date = datetime.now() - timedelta(days=random.randint(0, 3650))
    age = random.randint(23, 45)
    education = random.choice(education_map[department])
    performance_score = round(random.uniform(3.0, 5.0), 1)
    attrition = random.choice(["Yes", "No"])
    location = random.choice(locations)

    exit_date = None
    if attrition == "Yes":
        days_worked = (datetime.now() - hire_date).days
        if days_worked > 30:
           exit_date = hire_date + timedelta(days=random.randint(30, days_worked))
           exit_date = exit_date.date()
    else:
        attrition = "No" 

    data.append([
        emp_id, name, gender, department, role, salary,
        hire_date.date(), exit_date, age, education,
        performance_score, attrition, location
    ])

columns = [
    "employee_id", "name", "gender", "department", "role",
    "salary", "hire_date", "exit_date", "age",
    "education", "performance_score", "attrition", "location"
]

df_new = pd.DataFrame(data, columns=columns)

df_new.to_csv("employee_data_600_rows.csv", index=False)

print("600 rows generated successfully.")
```

```python
df_new['attrition_binary']= df_new['attrition'].map({'Yes':1,'No':0})
df_new.head()
```

```python
baseline_attrition=df_new['attrition_binary'].mean()
baseline_attrition
```

```python
df_new.groupby('location')['attrition_binary'].mean().sort_values(ascending=True) # banglore
df_new.groupby('education')['attrition_binary'].mean().sort_values(ascending=True) # MBA
df_new.groupby('age')['attrition_binary'].mean().sort_values(ascending=True) # 39
df_new.groupby('gender')['attrition_binary'].mean().sort_values(ascending=True) # female
```

```python
risky_stuff=((df_new['location']=='Bangalore')&(df_new['education']=='MBA'))
risky=df_new[risky_stuff]
risky.head()
others=df_new[~risky_stuff]
others
```

```python
risk_rate=risky['attrition_binary'].mean()
```

```python
others_rate=others['attrition_binary'].mean()
```

```python
risk_multiplier=risk_rate/others_rate
risk_multiplier
```

```python
round(risk_multiplier,2)
```
